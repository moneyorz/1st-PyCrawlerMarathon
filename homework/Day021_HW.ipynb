## 作業目標

根據範例：

* 取出今天所有的新聞
* 取出今天下午三點到五點的新聞
* 根據範例，取出三天前下午三點到五點的新聞

## 根據範例，取出今天所有的新聞

```python
from selenium import webdriver
from bs4 import BeautifulSoup
from selenium.webdriver.support.ui import Select
import time,re
browser = webdriver.Chrome(executable_path='chromedriver')
browser.get("https://www.ettoday.net/news/news-list.htm")
for i in range(30):
    time.sleep(2)
    browser.execute_script("window.scrollTo(0, 30000);")
print('OK')


```

```python
html_source = browser.page_source
soup = BeautifulSoup(html_source, "html5lib")
li1=list()
for d in soup.find(class_="part_list_2").find_all('h3'):
    if(re.match('2019/12/28 \d\d:\d\d',d.find(class_="date").text)):
        li1.append([d.find(class_="date").text,d.find_all('a')[-1].text])
for x in li1:
    print(x[0],x[1])
```

## 取出今天下午三點到五點的新聞

```python
li2=list()
for d in soup.find(class_="part_list_2").find_all('h3'):
    if(re.match('2019/12/28 (1[5-6]:\d\d|17:00)',d.find(class_="date").text)):
        li2.append([d.find(class_="date").text,d.find_all('a')[-1].text])
for x in li2:
    print(x[0],x[1])
```

## 根據範例，取出三天前下午三點到五點的新聞

```python
selectYear = Select(browser.find_element_by_id("selY"))
selectYear.select_by_value('2019')
selectMonth = Select(browser.find_element_by_id("selM"))
selectMonth.select_by_value('12')
selectDay = Select(browser.find_element_by_id("selD"))
selectDay.select_by_value('25')
browser.find_element_by_id('button').click()

for i in range(30):
    time.sleep(2)
    browser.execute_script("window.scrollTo(0, 30000);")

soup = BeautifulSoup(browser.page_source, "html5lib")

li3=list()

for d in soup.find(class_="part_list_2").find_all('h3'):
    if(re.match('2019/12/25 (1[5-6]:\d\d|17:00)',d.find(class_="date").text)):
        li3.append([d.find(class_="date").text,d.find_all('a')[-1].text])
for x in li3:
    print(x[0],x[1])
```
